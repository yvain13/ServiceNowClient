# Getting Started with the ServiceNow Python Client

## Prerequisites
Before diving in, ensure that you have the necessary installations and access to a ServiceNow instance. If you don’t have access to a development instance, a **Personal Developer Instance (PDI)** is a great starting point for experimentation.

---

## The ServiceNow Client

The **ServiceNowClient** serves as your interface to interact with a ServiceNow instance. Let’s walk through the basics.

### Step 1: Import the Library
First, import the necessary module:

```python
from pysnc import ServiceNowClient
```

### Step 2: Create a Client
To create a client, you can authenticate using the `ServiceNowPasswordGrantFlow`:

```python
client = ServiceNowClient(
    'dev00000',
    ServiceNowPasswordGrantFlow('admin', password, client_id, client_secret)
)
```

Alternatively, you can use a simpler but less secure method:

```python
client = ServiceNowClient('dev00000', ('admin', password))
```

---

## Using the Client

Once the client is created, you can use it to interact with ServiceNow records. For example, to work with the **Incident** table:

### Step 3: Create a `GlideRecord` Object
```python
gr = client.GlideRecord('incident')
```

This should feel familiar if you’ve used `GlideRecord` in other contexts. The key difference is that Python naming conventions are followed here.

### Step 4: Add Queries
Add queries to filter your results. For instance:

```python
gr.add_query("active", "true")
gr.query()
```

Or, use shorthand methods for common queries:

```python
gr.add_active_query()
```

````python
# Visualization functions
def create_visualizations(df):
    # Priority Distribution
    priority_counts = df["priority__display"].value_counts().reset_index()
    priority_counts.columns = ["priority", "count"]
    fig_priority = px.bar(
        priority_counts,
        x="priority",
        y="count",
        title="Ticket Priority Distribution",
        labels={"priority": "Priority", "count": "Count"},
    )

    # Category Distribution
    fig_category = px.pie(
        df,
        names="category__display",
        title="Category Distribution",
    )

    # Impact vs. Urgency
    fig_impact_urgency = px.scatter(
        df,
        x="impact__value",
        y="urgency__value",
        color="priority__display",
        size=[1]*len(df),
        title="Impact vs. Urgency",
        labels={"impact__value": "Impact", "urgency__value": "Urgency"},
    )

    # Combine all visuals in a single row
    fig = make_subplots(
        rows=1, cols=3,
        subplot_titles=("Priority Distribution", "Category Distribution", "Impact vs. Urgency"),
        specs=[[{"type": "xy"}, {"type": "domain"}, {"type": "scatter"}]]
    )

    # Add Priority Bar
    for trace in fig_priority.data:
        fig.add_trace(trace, row=1, col=1)

    # Add Category Pie
    for trace in fig_category.data:
        fig.add_trace(trace, row=1, col=2)

    # Add Impact vs. Urgency Scatter
    for trace in fig_impact_urgency.data:
        fig.add_trace(trace, row=1, col=3)

    # Update layout
    fig.update_layout(height=500, width=1500, title_text="Ticket Analysis Dashboard")
    fig.show()


# Run the functions

create_visualizations(df)
````
